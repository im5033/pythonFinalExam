```python
'''
請寫Python程式，引用 datetime 模組，印出今天的日期，並以民國年來顯示。
'''
import datetime as dt
now = dt.datetime.now()
year = int(now.year - 1911)
month = int(now.month)
day = int(now.day)
print(f"現在是民國{year}年{month}月{day}日")
```

```python
'''
請寫Python程式印出本課程在此學期中，所有18週上課的日期。
'''
import datetime as dt
first = dt.datetime(2022,9,14,0,0,0,0)
for i in range(18) :
    attend = first + dt.timedelta(weeks=i)
    print(f"第{i + 1}週上課是民國{attend.year - 1911}年{attend.month}月{attend.day}日")
```

```python
'''
請寫Python程式將 figures 目錄中的所有檔案，依其副檔名建立同名的目錄(以小寫表示)，
接著將這些檔案，分別搬移到對應的目錄中。
例如，a.jpg -> jpg 目錄，b.gif -> gif 目錄，其餘依此類推。
'''
import os
import shutil
dirs = []
cwd = os.getcwd() + "\\figures"
fds = os.listdir(cwd)
for f in fds:
    items = f.split('.') 
    folder = items[-1]  
    if os.path.isdir(cwd + "\\" + folder) == False:
        os.mkdir(cwd + "\\" + folder)
    if os.path.isfile(cwd + "\\" + f):
        shutil.move(cwd + "\\" + f , cwd + "\\" + folder + "\\" + f)
```

```python
'''
請產生一萬筆統一發票號碼，並請跟下列中獎資料兌獎，
並統計對獎的結果，也就是計算沒對中的跟對中1到8碼的各有多少？
'''
import random
win = [0,0,0,0,0,0,0,0,0]
na = "59276913"
seed = "1234567890"
for g in range(10000):
    tmp = 0
    list1 = []
    for h in range(8):
        list1.append(random.choice(seed))
    num = ''.join(list1)
    for i in range(8):
        if num[-8+i:] == na[-8+i:]:
            win[i] += 1
            break
        else:
            tmp += 1
        if tmp == 8:
            win[8] += 1
for j in range(9):
    print(f"對中{8-j}碼：{win[j]}")
```

```python
'''
1A2B遊戲
'''
from random import randint
secret = []
result = ''
times = 0
while len(secret) < 4:
    x = randint(0,9)
    tmp = 0
    if len(secret) > 0:
        for i in secret:
            if x != i : tmp += 1
        if tmp == len(secret) : secret.append(x)
    else : secret.append(x)
#print(secret)
while True:
    guess = input("Please input a 4-digit number : ")
    if guess.isdigit() and len(guess) == 4 :
        guess = [int(x) for x in guess]
        times += 1
        ans = ''
        a = 0
        b = 0
        for i in range(4):
            for j in range(4):
                if guess[i] == secret[j]:
                    if i == j:
                        a += 1
                    else:
                        b += 1
        if a>0 :
            ans += f'{a}A'
        if b>0 :
            ans += f'{b}B'
        if a==0 and b==0 :
            ans = '0'
        result += f'{times:02}: {guess} -> {ans}\n'
        print(result)
        if(ans == '4A'):
            print("Congratulation! You've got the number!")
            break
    else:
        print("You input not a 4-digit number.")
```

```python
'''
統計人名或姓氏出現的次數，並排序
'''
fp = open('Names-50.txt', "r")
name_str = fp.read()
fp.close()
names = name_str.split(", ")
last_name  = {}
first_name = {}
for name in names:
    last=name[0:int(abs(len(name)-2.5)+0.5)]
    first=name[len(last):len(name)]
    if last in last_name:
        last_name[last] += 1
    else:
        last_name[last] = 1
    if first in first_name:
        first_name[first] += 1
    else:
        first_name[first] = 1
print(sorted(last_name.items(), key=lambda x:x[1], reverse=True)) 
print(sorted(first_name.items(), key=lambda x:x[1], reverse=True)) 
```
